# using min max scaler to normalize our dataset to common scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df.drop(columns=['date', 'lights'])), 
                             columns=df.drop(columns=['date', 'lights']).columns)
features_df = normalised_df.drop(columns=['Appliances'])
df_target = normalised_df.Appliances
```

# Spliting data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, df_target, test_size=0.3, random_state=42)
```

## Fiting the model to train dataset

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train[['T2']], x_train.T6)
```

```python
T6_predicted = linear_model.predict(x_test[['T2']])
T6_predicted
```

## MAE 

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(x_test.T6,T6_predicted)
print (f'MAE: {round(mae,3)}')
```

## Residual Sum of Squares (RSS) 

```python
import numpy as np
from sklearn.metrics import mean_squared_error
rss = np.sum(np.square(x_test.T6-T6_predicted))
print(f'Residual Sum of Squares (RSS): {round(rss,3)}')
```

## Root Mean Square Error (RMSE)

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(x_test.T6, T6_predicted))
print ( f'Root Mean Square Error (RMSE): {round(rmse,3)}')
```

## R-Squared 

```python
from sklearn.metrics import r2_score
r2_sqr = r2_score(x_test.T6 , T6_predicted)
print(f'R-Squared: {round(r2_sqr,3)}')
```

## Ridge Regression 

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

## Feature Selection and Lasso Regression 

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

## comparing the effects of regularisation


```python
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model.fit(x_train,y_train)
```

```python

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')

# linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
# ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
# lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

# final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
# final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
